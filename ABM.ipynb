```python
import numpy as np
import time
import matplotlib.pyplot as plt
import random
import time
```

### Bennati model

```python
def bennati_rule(x, y, dx):
    s = random.randint(0, 1)
    if (x + s*dx - (1-s)*dx) > 0 and (y + (1-s)*dx - s*dx) > 0:
        return (x + s*dx - (1-s)*dx), (y + (1-s)*dx - s*dx)
    else:
        return x, y
```

```python
# parameters
N = 100
n_cycles = 1000
dx = 0.1

x = np.ones(N)
start = time.time()
for n in range(n_cycles):
    for i in range(N-1):
        for j in range(i+1, N):
            
            x[i],x[j] = bennati_rule(x[i], x[j], dx)
            
end = time.time()
run_time = end - start
print(f'run time = {run_time} seconds')

# hist
plt.hist(x, bins=int(np.sqrt(N)), density=True, label='empirical pdf')

plt.legend()
plt.title('wealth distribution')
plt.show()
```

### Dragulescu model

```python
def dragulescu_rule(x, y):
    eps = random.random()
    return eps*(x+y), (1-eps)*(x+y)
```

```python
# parameters
N = 100
n_cycles = 1000

x = np.ones(N)
start = time.time()
for n in range(n_cycles):
    for i in range(N-1):
        for j in range(i+1, N):
            x[i],x[j] = dragulescu_rule(x[i], x[j])
            
end = time.time()
run_time = end - start
print(f'run time = {run_time} seconds')

# exp pdf
xx = np.linspace(0, 5, 1000)
ff = (1 / np.mean(x)) * np.exp(-xx/np.mean(x))

# charts
plt.hist(x, bins=int(np.sqrt(N)), density=True, label='empirical pdf')
plt.plot(xx, ff, label='exponential pdf')

plt.legend()
plt.title('wealth distribution')
plt.show()
```

### Chakraborti model

```python
def chakraborti_rule(x, y, lambda_):
    eps = random.random()
    return x*lambda_ + eps*(1-lambda_)*(x+y), y*lambda_ + (1-eps)*(1-lambda_)*(x+y)
```

```python
# parameters
N = 100
n_cycles = 1000
lambda_ = 0.5

x = np.ones(N)
start = time.time()
for n in range(n_cycles):
    for i in range(N-1):
        for j in range(i+1, N):
            x[i], x[j] = chakraborti_rule(x[i], x[j], lambda_)
            
end = time.time()
run_time = end - start
print(f'run time = {run_time} seconds')

# hist
plt.hist(x, bins=int(np.sqrt(N)), density=True, label='empirical pdf')

plt.legend()
plt.title('wealth distribution')
plt.show()
```

### Chatterjee model

```python
def chatterjee_rule(x, y, lambda_1, lambda_2):
    eps = random.random()
    x_p = x*lambda_1 + eps*((1-lambda_1)*x + (1-lambda_2)*y)
    y_p = y*lambda_2 + (1-eps)*((1-lambda_1)*x + (1-lambda_2)*y)
    
    return x_p, y_p
```

```python
# parameters
N = 100
n_cycles = 1000
lambda_ = np.random.rand(N)

x = np.ones(N)
start = time.time()
for n in range(n_cycles):
    for i in range(N-1):
        for j in range(i+1, N):
            x[i], x[j] = chatterjee_rule(x[i], x[j], lambda_[i], lambda_[j])
            
end = time.time()
run_time = end - start
print(f'run time = {run_time} seconds')

# hist
plt.hist(x, bins=int(np.sqrt(N)), density=True, label='empirical pdf')

plt.legend()
plt.title('wealth distribution')
plt.show()
```
